#Importação das bibliotecas

```python
import pandas as pd
import numpy as np
```

#Leitura das bases de bots e humanos

Abro todos os conjuntos de dados, os 3 conjuntos com bots e 1 conjunto com humanos

```python
df_spambots1 = pd.read_csv("df_bots_spam_1.csv",sep =";")
df_spambots2 = pd.read_csv("df_bots_spam_2.csv",sep =";")
df_spambots3 = pd.read_csv("df_bots_spam_3.csv",sep =";")
df_humans = pd.read_csv("df_userslegitims.csv",sep =";")
```

Verificação da quantidade de humanos

```python
df_humans.shape
```

#Seleção aleatória de amostras das bases com bots

Separo amostras de cada conjunto de acordo com a seguinte lógica, 1083/3 = 361
que é a quantidade de amostras de humanos (1083) dividido pela quantidade de bases com bots (3)

```python
df_s1 = df_spambots1.sample(n=361)
df_s2 = df_spambots2.sample(n=361)
df_s3 = df_spambots3.sample(n=361)
```

Verifico a quantidade de amostras

```python
df_s1.shape
```

#União de todas as amostras com a base de humanos

```python
df = pd.concat([df_humans,df_s1,df_s2,df_s3], ignore_index=True)
```

Verifico a quantidade de amostras e colunas do novo conjunto

```python
df.shape
```

#Distribuição de humanos e bots na base

Verifico se a quantidade de bots e humanos estão iguais no novo conjunto

```python
df.tipo_usuario.value_counts().plot(kind='pie') #0 é humano e 1 é bot
```

#Exportando o conjunto de dados

```python
df.to_csv('df_hibrido.csv',sep = ';', line_terminator = '', 
                     encoding = 'utf-8', index=False) # no fim exporto um csv com uma base pré-montada
```
